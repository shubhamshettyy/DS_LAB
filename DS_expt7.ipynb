```python
#Build a new corpus variable called corpus
new_corpus = df.copy()
```

```python
#convert the text to lowercase. 
for i in range(len(new_corpus)):
  new_corpus['text'][i] = new_corpus['text'][i].lower()
new_corpus.head()
```

### Remove punctuation from emails

### Removing stop words

```python
from nltk.tokenize import word_tokenize

for i in range(len(new_corpus)):
  new_corpus['text'][i] = word_tokenize(new_corpus['text'][i])
  new_corpus["text"][i] = [w for w in new_corpus["text"][i] if not w.lower() in stop_words_maam]

new_corpus.head()  
```

### Stemming the emails using SnowballStemmer

```python
#Stemming

from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer

ps = PorterStemmer()
sb = SnowballStemmer("english")

for i in range(len(df)):
    arr = []
    for ele in new_corpus['text'][i]:
        ele = sb.stem(ele)
        arr.append(ele)
    new_corpus['text'][i] = arr
```

### Creating the Document Term Matrix

```python
import textmining

# Initialize class to create term-document matrix
tdm = textmining.TermDocumentMatrix()

for i in range(len(new_corpus)):
    tdm.add_doc(new_corpus['text'][i])
```

```python
tdm_df = tdm.to_df(cutoff=0)
tdm_df
```

### A threshold of 5% is applied

```python
#including terms which appear in 5% 
threshold = 286
sptdm = tdm.to_df(cutoff=threshold)
```

## Building various models

### Creating a test-train split

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

Y = new_corpus['spam']

X_train, X_test, y_train, y_test = train_test_split(sptdm, Y, test_size = 0.3, random_state = 123)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

### CART classifier

```python
from sklearn import tree

cart = DecisionTreeClassifier()
cart.fit(X_train, y_train)
```

```python
print(f"training score: {cart.score(X_train, y_train)}")
```

```python
THRESHOLD = 0.5
y_pred = np.where(cart.predict_proba(X_test)[:,1] >= THRESHOLD, 1, 0)
print(f"testing score: {accuracy_score(y_test, y_pred)}")
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

spamRF = RandomForestClassifier(random_state=123)
spamRF.fit(X_train, y_train)
```

```python
print(f"training score: {spamRF.score(X_train, y_train)}")
```

```python
THRESHOLD = 0.5
y_pred = np.where(spamRF.predict_proba(X_test)[:,1] >= THRESHOLD, 1, 0)
print(f"testing score: {accuracy_score(y_test, y_pred)}")
```

### AUC Score

#### Train set

```python
from sklearn import metrics
from sklearn.metrics import roc_auc_score
predCart = cart.predict(X_train)
predRF = spamRF.predict(X_train)
print(f"Train set cart AUC: {roc_auc_score(y_train, predCart)}")
print(f"Train set RF AUC: {roc_auc_score(y_train, predRF)}")
```

#### Train set with threshold

```python
THRESHOLD = 0.5
y_pred_cart = np.where(cart.predict_proba(X_train)[:,1] >= THRESHOLD, 1, 0)
y_pred_RF = np.where(spamRF.predict_proba(X_train)[:,1] >= THRESHOLD, 1, 0)
print(f"Train set cart with threshold AUC: {roc_auc_score(y_train, y_pred_cart)}")
print(f"Train set RF with threshold AUC: {roc_auc_score(y_train, y_pred_RF)}")
```

#### Test set

```python
predCart = cart.predict(X_test)
predRF = spamRF.predict(X_test)
print(f"Test set cart AUC: {roc_auc_score(y_test, predCart)}")
print(f"Test set RF AUC: {roc_auc_score(y_test, predRF)}")
```

```python
THRESHOLD = 0.5
y_pred_cart = np.where(cart.predict_proba(X_test)[:,1] >= THRESHOLD, 1, 0)
y_pred_RF = np.where(spamRF.predict_proba(X_test)[:,1] >= THRESHOLD, 1, 0)

print(f"Test set cart with threshold AUC: {roc_auc_score(y_test, y_pred_cart)}")
print(f"Test set RF with threshold AUC: {roc_auc_score(y_test, y_pred_RF)}")
```

## Conclusion


*   A spam classifier can potentially benefit from including the frequency of the word that appears in every email. (Document term matrix)
*   Random Forest model outperformed CART model in terms of Accuracy and AUC.
*   Random forests outperform Cart because they consist of multiple single trees each based on a random sample of the training data. They are typically more accurate than single decision trees.
*   While a single decision tree like CART is often pruned, a random forest tree is fully grown and unpruned.
*   Each random forest tree is learned on a random sample, and at each node, a random set of features are considered for splitting. Both mechanisms create diversity among the trees which in turn give sus more accurate results.
